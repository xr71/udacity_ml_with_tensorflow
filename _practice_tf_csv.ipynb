```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf 

```

```python
df = pd.read_csv("testdata.csv", header=None)
```

```python
plt.scatter(df[0], df[1])
```

```python
X, y = df[0], df[1]
```

```python
nn = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])
])
```

```python
nn.summary()
```

```python
nn.compile(optimizer="adam", loss="mse", metrics=["mse"])
```

```python
hist = nn.fit(X, y, epochs=1000)
```

```python
nn.get_weights()
```

```python
plt.scatter(X, y)
plt.plot(X, nn.predict(X))
```
